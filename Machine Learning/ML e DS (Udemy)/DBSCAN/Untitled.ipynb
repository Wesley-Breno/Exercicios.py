<h1>Importing the necessary data</h1>

```python
import numpy as np
import plotly.express as px
from sklearn.preprocessing import StandardScaler
```

<h1>Defining the data that will be used to perform the grouping</h1>

```python
x = [20, 27, 21, 37, 46, 53, 55, 47, 52, 32, 39, 41, 39, 48, 48]
y = [1000, 1200, 2900, 1850, 900, 950, 2000, 2100, 3000, 5900, 4100, 5100, 7000, 5000, 6500]
```

```python
base = np.array(
    [
        [20, 1000],
        [27, 1200],
        [21, 2900],
        [37, 1850],
        [46, 900],
        [53, 950],
        [55, 2000],
        [47, 2100],
        [52, 3000],
        [32, 5900],
        [39, 4100],
        [41, 5100],
        [39, 7000],
        [48, 5000],
        [48, 6500]
    ]
)
```

```python
base
```

<h1>Scalling data</h1>

```python
scaler = StandardScaler()
base_scaled = scaler.fit_transform(base)
```

```python
base_scaled
```

<h1>Creating clustering with DBSCAN</h1>

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.95, min_samples=2)
dbscan.fit(base_scaled)
```

<h1>Showing grouping results</h1>

```python
labels = dbscan.labels_
labels
```

```python
grafic = px.scatter(x = base[:,0], y = base[:,1], color=labels)
grafic.show()
```
