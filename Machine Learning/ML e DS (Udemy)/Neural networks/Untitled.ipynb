<h1>Separating predictive data from classifiers</h1>

```python
x_credit = df.iloc[:, 1:4].values
x_credit
```

```python
y_credit = df.iloc[:, 4].values
y_credit
```

<h1>Scaling data</h1>

```python
from sklearn.preprocessing import StandardScaler
scaler_credit = StandardScaler()
x_credit = scaler_credit.fit_transform(x_credit)
```

<h1>Separating training data and testing data</h1>

```python
from sklearn.model_selection import train_test_split

x_credit_train, x_credit_test, y_credit_train, y_credit_test = train_test_split(
    x_credit,  
    y_credit,
    test_size=0.25, # Splitting the data into 25% for testing and the rest for training
    random_state=0  # Controls randomness in the division of data (In this case there is no randomness)
)
```

```python
x_credit_train.shape, x_credit_test.shape
```

```python
y_credit_train.shape, y_credit_test.shape
```

<h1>Training the algorithm with separate data</h1>

```python
from sklearn.neural_network import MLPClassifier
```

```python
neural_network = MLPClassifier()
neural_network.fit(x_credit_train, y_credit_train)
```

```python
neural_network = MLPClassifier(max_iter=1000, verbose=True)
neural_network.fit(x_credit_train, y_credit_train)
```

```python
neural_network = neural_network = MLPClassifier(max_iter=1000, verbose=True, tol=0.0000100)
neural_network.fit(x_credit_train, y_credit_train)
```

```python
neural_network = neural_network = MLPClassifier(
    max_iter=1500, 
    verbose=True, 
    tol=0.0000100, 
    solver='adam', 
    activation='relu', 
    hidden_layer_sizes=(2, 2)
)
neural_network.fit(x_credit_train, y_credit_train)
```

<h1>Showing model performance</h1>

```python
predicts = neural_network.predict(x_credit_test)
```

```python
predicts
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_credit_test, predicts)
```

```python
print(classification_report(y_credit_test, predicts))
```
